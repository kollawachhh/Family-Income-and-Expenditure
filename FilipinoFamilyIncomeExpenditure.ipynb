ในที่นี่เราสนใจ 33 คอลัมป์แรกที่เกี่ยวกับค่าใช้จ่ายและข้อมูลพื้นฐานของหัวหน้าครอบครัว

```python
fie = fie.iloc[:,:33]
fie["Main Source of Income"] = pd.Categorical(fie["Main Source of Income"])
fie["Region"] = pd.Categorical(fie["Region"])
fie["Household Head Marital Status"] = pd.Categorical(fie["Household Head Marital Status"])
fie["Household Head Job or Business Indicator"] = pd.Categorical(fie["Household Head Job or Business Indicator"])
fie["Type of Household"] = pd.Categorical(fie["Type of Household"])
fie.drop(["Household Head Occupation","Household Head Class of Worker"],axis=1,inplace=True)
fie
```

## 3.1 รายได้เฉลี่ยของแต่ละภูมิภาคและประเภทรายได้ที่หักด้วยรายจ่ายเฉลี่ย

```python
expenditureCol= []
for i in fie.columns:
    if ("Expenditure" in i and i != "Total Food Expenditure") :
        expenditureCol.append(i)
expenditureCol
```

```python
fieTotalIncome = fie.loc[:,"Total Household Income"]
fieTotalExpen = fie.loc[:,expenditureCol].sum(axis=1)
fieIncomeExpenDiff = pd.DataFrame(fieTotalIncome - fieTotalExpen).rename(columns={0:"NetIncome"})
fieIncomeExpenDiff = fieIncomeExpenDiff.join(fie.Region)
fieIncomeExpenDiff = fieIncomeExpenDiff.join(fie["Main Source of Income"])
fieIncomeExpenDiff = fieIncomeExpenDiff.groupby(["Region","Main Source of Income"]).mean()
fieIncomeExpenDiff
```

```python
sns.catplot(data=fieIncomeExpenDiff.reset_index(), y='NetIncome',x='Region', col="Main Source of Income",kind='bar').set_xticklabels(rotation=90).set(ylim=(0, 200000))
```

## 3.3 ประเภทของรายจ่ายเฉลี่ยที่มากที่สุดในแต่ละภูมิภาค

```python
fieExpen = fie.loc[:,expenditureCol]
fieExpen = fieExpen.join(fie.Region)
fieExpen = fieExpen.groupby(["Region"]).mean()
fieExpen
```

```python
longFieExpen = fieExpen.reset_index()
longFieExpen = longFieExpen.melt(id_vars=["Region"])
longFieExpen = longFieExpen.rename(columns={"variable":"ExpenditureType"})
longFieExpen = longFieExpen.set_index(["Region","ExpenditureType"])
longFieExpen = longFieExpen.sort_values("value",ascending=False).sort_index(level=0,sort_remaining=False)
longFieExpen
```

```python
maxFieExpenType = longFieExpen.reset_index()
maxFieExpenType = maxFieExpenType.drop_duplicates("Region")
maxFieExpenType = maxFieExpenType.set_index("Region")
maxFieExpenType["ExpenditureType"] = pd.Categorical(maxFieExpenType["ExpenditureType"])
maxFieExpenType
```

```python
sns.catplot(data=maxFieExpenType.reset_index(), x='Region', y='value',hue='ExpenditureType', kind='bar').set_xticklabels(rotation=90)
```
