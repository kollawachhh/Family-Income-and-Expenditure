## รายละเอียดข้อมูล ##

**คำอธิบายข้อมูล**

**รายละเอียดคอลัมน์**
1. imbd_title_id    - รหัสของหนังแต่ละเรื่อง
2. title                 - ชื่อเรื่อง
3. original_title     - ชื่อเรื่องจากต้นฉบับ
4. year                - ปีที่กำหนดฉาย
5. date_published - วันที่กำหนดฉาย
6. genre              - ประเภทของหนัง
7. duration          - ระยะเวลาของหนัง(ต้นเรื่อง - จบเรื่อง)
8. country           - ประเทศที่ผลิต
9. language         - ภาษาของหนัง
10. director         - ผู้กำกับการแสดง
11. writer            - ผู้เขียน
12. production_company - บริษัทที่สร้างหนัง
13. actor             - นักแสดง
14. description     - คำอธิบายหนัง, เกริ่นเนื้อเรื่องย่อ
15. avg_vote       - ค่าเฉลี่ยคะแนนโหวต
16. votes             - จำนวนที่โหวตทั้งหมด
17. budget           - งบประมาณการสร้าง
18. usa_gross_income - รายได้รวมจากการฉายหนัง
19. worlwide_gross_income - รายได้รวมทั่วโลกจากการฉายหนัง
20. metascore       - คะแนนจากการวิจารณ์ของสื่อต่างๆ
21. reviews_from_users - การรีวิวจากผู้ชม
22. reviews_from_critics - การรีวิลจากนักวิจารณ์
