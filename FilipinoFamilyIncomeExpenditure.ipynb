## 3. การวิเคราะห์ข้อมูลเชิงสำรวจเบื้องต้น

```python
fie.shape
```

### สนใจเพียงคอลัมป์ที่เกี่ยวกับค่าใช้จ่ายและข้อมูลพื้นฐานของหัวหน้าครอบครัว

```python
fie = fie.iloc[:,:34]
```

### เปลี่ยนบางคอลัมน์เป็น Categories เพื่อให้สะดวกในการสร้างกราฟ

```python
fie["Main Source of Income"] = pd.Categorical(fie["Main Source of Income"])
fie["Region"] = pd.Categorical(fie["Region"])
fie["Household Head Marital Status"] = pd.Categorical(fie["Household Head Marital Status"])
fie["Household Head Job or Business Indicator"] = pd.Categorical(fie["Household Head Job or Business Indicator"])
fie["Type of Household"] = pd.Categorical(fie["Type of Household"])
```

```python
fie
```

### สร้าง expenditure_col ซึ่งเป็น List ที่เก็บ Column ต่างๆที่เป็นค่าใช้จ่ายทั้งหมด

```python
expenditure_col= []
for i in fie.columns:
    if ("Expenditure" in i and i != "Total Food Expenditure") :
        expenditure_col.append(i)
expenditure_col
```

## 4. ความน่าสนใจที่หาได้จากข้อมูลชุดนี้

### 4.1 รายได้เฉลี่ยหักด้วยรายจ่ายเฉลี่ยของ แยกตามภูมิภาคและประเภทรายได้

```python
fie_total_income = fie.loc[:,"Total Household Income"]
fie_total_expen = fie.loc[:,expenditure_col].sum(axis=1)
fie_income_expen_diff = pd.DataFrame(fie_total_income - fie_total_expen).rename(columns={0:"Balance"})
fie_income_expen_diff = fie_income_expen_diff.join(fie.Region)
fie_income_expen_diff = fie_income_expen_diff.join(fie["Main Source of Income"])
fie_income_expen_diff = fie_income_expen_diff.groupby(["Region","Main Source of Income"]).mean()
fie_income_expen_diff
```

```python
sns.catplot(data=fie_income_expen_diff.reset_index(), y='Balance',x='Region', 
            col="Main Source of Income",kind='bar').set_xticklabels(rotation=90).set(ylim=(0, 200000))
```

- รายได้เฉลี่ยหักลบรายจ่ายเฉลี่ยแล้ว ในภูมิภาค NCR มีเงินเฉลี่ยที่คงเหลือ(ฺBalance) มากที่สุด
- ครัวเรือนที่มีรายได้ Wage/Salaries จะมีเงินเฉลี่ยที่คงเหลือ(ฺBalance) ใกล้เคียงกันมากที่สุด

## 4.2 ค่าเฉลี่ยของรายได้ของแต่ละภูมิภาค แยกตามประเภทรายได้และเพศของหัวหน้าครอบครัว

```python
mean_income_region = fie.groupby(['Region','Main Source of Income','Household Head Sex']).mean()
mean_income_region = mean_income_region.loc[:,['Total Household Income']]
mean_income_region
```

```python
sns.catplot(data=mean_income_region.reset_index(),x='Region',y='Total Household Income',col='Main Source of Income',
    hue='Household Head Sex',kind='bar').set_xticklabels(rotation=90)
```

- รายได้เฉลี่ยส่วนใหญ่ในภูมิภาค NCR มากที่สุดในทุกประเภทของ
- ผู้คนส่วนใหญ่จะมีรายได้แบบ Wage/Salaries
- รายได้ไม่ผูกขาดอยู่กับเพศ บางส่วนเพศหญิงก็มีรายได้มากกว่าเพศชาย

## 4.3 ประเภทของรายจ่ายเฉลี่ยที่มากที่สุดในแต่ละภูมิภาค ##

```python
fie_expen = fie.loc[:,expenditure_col]
fie_expen = fie_expen.join(fie.Region)
fie_expen = fie_expen.groupby(["Region"]).mean()
fie_expen
```

```python
long_fie_expen = fie_expen.reset_index()
long_fie_expen = long_fie_expen.melt(id_vars=["Region"])
long_fie_expen = long_fie_expen.rename(columns={"variable":"ExpenditureType"})
long_fie_expen = long_fie_expen.set_index(["Region","ExpenditureType"])
long_fie_expen = long_fie_expen.sort_values("value",ascending=False)
long_fie_expen = long_fie_expen.sort_index(level=0,sort_remaining=False)
long_fie_expen
```

```python
max_fie_expen_type = long_fie_expen.sort_index(level=0,sort_remaining=False).reset_index()
max_fie_expen_type = max_fie_expen_type.drop_duplicates("Region")
max_fie_expen_type = max_fie_expen_type.set_index("Region")
max_fie_expen_type["ExpenditureType"] = pd.Categorical(max_fie_expen_type["ExpenditureType"])
max_fie_expen_type
```

```python
sns.catplot(data=max_fie_expen_type.reset_index(), x='Region', y='value',
            hue='ExpenditureType', kind='bar').set_xticklabels(rotation=90)
```

- ภูมิภาค NCR ค่าใช้จ่ายในหนึ่งประเภทของรายจ่ายที่มากที่สุด(ประเภทรายจ่ายของแต่ละภูมิภาค) สูงกว่าค่ารายจ่ายสูงสุดของภูมิภาคอื่นๆ
- ภูมิภาคส่วนใหญ่มีรายจ่ายที่มากที่สุดคือ Housing and water(ค่าน้ำและค่าไฟ) และรองลงมาคือ Bread and Cereals (ขนมปังและธัญพืช)
